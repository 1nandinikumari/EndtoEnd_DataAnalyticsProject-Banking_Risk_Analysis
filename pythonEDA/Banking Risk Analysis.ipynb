```python
#Total Balance
df['Total Balance'] = df[['Checking Accounts','Saving Accounts','Bank Deposits']].sum(axis=1)
df['Total Balance'].head()
```

```python
#Replace impossible ages
df.loc[(df['Age'] < 18) | (df['Age'] > 90), 'Age'] = np.nan
```

```python
###Checking and Replace negative values (just in case any exist)
cols_money = ['Bank Loans','Credit Card Balance','Bank Deposits','Saving Accounts','Checking Accounts']
for col in cols_money:
    negatives = df[df[col] < 0]
    print(col, " → negative rows:", len(negatives))
    if len(negatives) > 0:
        display(negatives.head())
```

```python
#Create Debt-to-Income Ratio

#It shows how much liability a customer has compared to income.

df['Debt_To_Income'] = (df['Bank Loans'] + df['Credit Card Balance']) / df['Estimated Income']
df['Debt_To_Income'].head()
```

```python
#Visualize Debt-to-Income distribution
plt.figure(figsize=(6,4))
sns.histplot(df['Debt_To_Income'], kde=True)
plt.title("Distribution of Debt to Income Ratio")
plt.xlabel("Debt to Income Ratio")
plt.show()
```

```python
#Flag High Risk Customers
#flag column: 1 = high risk, 0 = normal

df['High_Risk_Flag'] = np.where(df['Debt_To_Income'] > 0.6, 1, 0)
df['High_Risk_Flag'].value_counts()
```

```python
# High Risk vs Income (boxplot)
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='High_Risk_Flag', y='Estimated Income')
plt.title("Estimated Income by Risk Category")
plt.xlabel("Risk Category (0 = Normal, 1 = High Risk)")
plt.ylabel("Estimated Income")
plt.show()
```

### Estimated Income by Risk Category
From the boxplot, we can see that the High Risk customers generally have lower median income compared to Normal Risk customers. 
This implies that customers with lower income are more financially stressed and more likely to fall in High Risk category. Therefore, income is a key indicator for identifying risky customers.


```python
# Products Count Column (more products = more exposure)
product_cols = ['Checking Accounts','Saving Accounts','Bank Loans','Credit Card Balance']

df['Products_Count'] = df[product_cols].notna().sum(axis=1)
df['Products_Count'].head()
```

```python
# Bar Chart (how many customers have 1,2,3 products etc.)
plt.figure(figsize=(6,4))
df['Products_Count'].value_counts().sort_index().plot(kind='bar')
plt.title("How many products customers hold")
plt.xlabel("Number of Products")
plt.ylabel("Customer Count")
plt.show()
```

### 

Most of the customers hold 4 different banking products. This means majority of the customer base is deeply engaged with the bank and using multiple services like deposits, savings, cards and loans together.

Higher product ownership indicates higher dependency on the bank which is positive for cross-selling and revenue generation. However, customers with too many products should also be monitored as higher exposure across multiple products may also increase the overall credit risk.

## 📝 Final Project Summary – Banking Risk Analysis

In this risk analysis project, I performed data cleaning, transformation and exploratory data analysis on banking customer data to understand the financial stability and credit exposure of customers. I derived additional metrics such as **Debt-to-Income Ratio**, **Disposable Income** and **Products Count** to identify financially stressed customers and understand where risk is concentrated.

---

### 🔹 Risk Weighting Insights
From the risk weighting distribution (bar/pie charts), most customers fall under **Risk Weight 1 & 2** categories. This means the larger portion of the bank’s customer base is low-to-medium risk. Very few customers fall in higher risk categories (Risk 4 & 5), which indicates that overall the portfolio is stable, but a small high-risk group must be monitored more closely.

---

### 🔹 Income Band Insights
The Income Band visualization shows that **most customers are in the Medium and Low income categories**. Only a small part of the customers fall under the High income band. Lower income segments generally have lower financial buffers and are more prone to financial stress, which may increase risk in the future. This insight can help the bank plan stricter credit checks for lower income groups.

---

### 🔹 Products Ownership Insights
The Products Count chart shows that **majority of customers hold 4 products**. This indicates high product penetration and strong customer engagement. While this is beneficial for business growth and cross-selling, it also means customers are more exposed across multiple financial products. This makes it important to monitor customers with high product usage combined with high debt ratios.

---

### 🔹 Loyalty Classification vs Risk Weighting

From the chart, it is observed that customers in the Gold and Jade loyalty segments have a higher count in lower risk weighting categories (Risk 1 and 2). This indicates that these groups are relatively more financially stable and less likely to default.

On the other hand, Silver customers show a noticeable presence across higher risk categories (Risk 3, 4, and 5). This suggests that the Silver segment contains comparatively riskier customers who may have unstable financial behaviour and require more monitoring.

Overall, loyalty classification is a good indicator of risk. Higher loyalty tiers (Gold, Jade, Platinum) are more stable and lower loyalty tiers (especially Silver) represent customer groups where probability of risk is higher.

---

### ✅ Conclusion
Overall, the bank’s customer portfolio appears to be primarily low-to-medium risk with stable income patterns and high product usage. However, **income-sensitive customers and certain occupations** represent higher risk profiles. Targeted monitoring, early warning triggers and focused credit management strategies for these smaller high-risk segments can help improve risk governance and reduce potential defaults.
